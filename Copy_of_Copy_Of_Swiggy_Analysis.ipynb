New Prediction For Average Ratings

```
import pandas as pd
import joblib

from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor

from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score


x=df.drop(["Avg ratings","Restaurant","Address","Area","City","ID"], axis=1)
y=df["Avg ratings"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

num=x.select_dtypes(include=["int64","float64"]).columns
cat=x.select_dtypes(include="object").columns

n_pipeline=Pipeline([
    ("imputer",SimpleImputer(strategy="mean")),
    ("scaler",StandardScaler())
])

c_pipeline=Pipeline([
    ("onehot",OneHotEncoder(handle_unknown="ignore"))
])

preprocess=ColumnTransformer([
    ("num",n_pipeline,num),
    ("cat",c_pipeline,cat)
])

model_dict=({

             "Decision":DecisionTreeRegressor(),
             "Random":RandomForestRegressor(),
             "Gradient":GradientBoostingRegressor(),
             "XGB":XGBRegressor()
})

result=[]
for name,model in model_dict.items():
  pipeline=Pipeline([
      ("Preprocesser",preprocess),
      ("Models",model)
  ])

  pipeline.fit(x_train,y_train)
  y_pred=pipeline.predict(x_test)

  mae=mean_absolute_error(y_test,y_pred)
  mse=mean_squared_error(y_test,y_pred)
  rmse=root_mean_squared_error(y_test,y_pred)
  r2=r2_score(y_test,y_pred)

  train_score=pipeline.score(x_train,y_train)
  test_score=pipeline.score(x_test,y_test)

  if train_score <0.80 and test_score < 0.80:
   status= "Underfit"
  elif train_score -test_score >1.10:
    status="Overfit"
  else:
    status="Good"


  result.append([name,mae,mse,rmse,r2,train_score,test_score,status])
result_df=pd.DataFrame(result,columns=["Model","MAE","MSE","RMSE","R2_Score","Train_Score","Test_Score","Model_Status"])
print(result_df.sort_values(by="R2_Score",ascending=False))

params=(

        {
            "Models":[GradientBoostingRegressor()],
            "Models__n_estimators":[30,50,80],
            "Models__max_depth":[None,5,8],
            "Models__learning_rate":[0.02,0.08,0.1]
        }
)

grid=GridSearchCV(
    pipeline,
    params,
    cv=5,
    scoring="r2",
    verbose=3,
    n_jobs=-1
)

grid.fit(x_train,y_train)
best_model=grid.best_estimator_



new_restaurant = x.iloc[[0]].copy()
new_restaurant["City"]= 1,
new_restaurant["Price"]= 600,
new_restaurant["Total Ratings"]= 2500,
new_restaurant["Delivery Time"]= 25,
new_restaurant["Delivery Status"]= 1,
new_restaurant["Thai"]= 0,
new_restaurant["Thai"]= 0,
new_restaurant["Thai"]= 0,
new_restaurant["Thai"]= 0,
new_restaurant ["Thalis"]= 0,
new_restaurant["Tibetan"]= 0,
new_restaurant["Tibetan"]= 0,
new_restaurant["Tibetan Chinese"]= 0,
new_restaurant["Asian"]= 0,
new_restaurant ["Snacks"]= 1,
new_restaurant["Continental"]= 0,
new_restaurant["Desserts"]= 1,
new_restaurant["Beverages"]= 1,
new_restaurant["Tribal"]= 0,
new_restaurant["Turkish"]= 0,
new_restaurant["Use code Xpress121 to avail"]= 0,
new_restaurant["Vietnamese"]= 0,
new_restaurant["Waffle"]= 0,
new_restaurant["Indian"]= 1


prediction = best_model.predict(new_restaurant)

print("Predicted Avg Rating:", prediction[0])

print(prediction)

joblib.dump(pipeline,"Swiggy_model.pkl")

loaded_model=joblib.load("Swiggy_model.pkl")

print("Pipeline Saved and Loaded Successfully")
```

```
import pandas as pd
import joblib
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

x=df[["Area","City","Address","Restaurant"]]
y=df["DeliveryStatus"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

num=x.select_dtypes(include=["int64","float64"]).columns
cat=x.select_dtypes(include="object").columns

n_pipeline=Pipeline([
    ("imputer",SimpleImputer(strategy="mean")),
    ("scaler",StandardScaler())
])

c_pipeline=Pipeline([
    ("encoder",OneHotEncoder(handle_unknown="ignore"))
])

preprocess=ColumnTransformer([
    ("num",n_pipeline,num),
    ("cat",c_pipeline,cat)
])

model_dict=({

             "Decision":DecisionTreeClassifier(),
             "Random":RandomForestClassifier(),
             "Gradient":GradientBoostingClassifier(),
             "XGB":XGBClassifier(),
             "KNN":KNeighborsClassifier()
})

result=[]
for name,model in model_dict.items():
  pipeline=Pipeline([
    ("Preprocessor",preprocess),
    ("Models",model)
])

  pipeline.fit(x_train,y_train)
  y_pred=pipeline.predict(x_test)


  acc=accuracy_score(y_test,y_pred)
  pre=precision_score(y_test,y_pred, average='weighted')
  re=recall_score(y_test,y_pred, average='weighted')
  f1=f1_score(y_test,y_pred, average='weighted')

  train_score=pipeline.score(x_train,y_train)
  test_score=pipeline.score(x_test,y_test)

  if train_score < 0.80 and test_score < 0.80:
   status="Underfit"
  elif train_score - test_score >1.10:
   status="Overfit"
  else:
   status="Good"

  result.append([name,acc,pre,re,f1,train_score,test_score,status])
result_df=pd.DataFrame(result,columns=["Model","Accuracy","Precision","Recall","F1_Score","Train_Score","Test_Score","Status"])
print(result_df.sort_values(by="Accuracy",ascending=False))

params=(

        {
            "Model":[XGBClassifier()],
            "Model__n_estimators":[30,50,80],
            "Model__max_depth":[None,3,6]
        }
)

grid=GridSearchCV(
    pipeline,
    params,
    cv=5,
    scoring="accuracy",
    n_jobs=-1,
    verbose=3
)

grid.fit(x_train,y_train)
best_model=grid.best_estimators_


new=x.iloc[[0]]
new["Area"]= 789,
new["City"]= 18,
new["Address"]= 11,
new["Restaurant"]=16

prediction=best_model.predict(new)
print("New Prediction Del_Status")
print(prediction)
```
